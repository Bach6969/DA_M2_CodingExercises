```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
"genre": [
    "Pop","Rock","Hip-Hop","Electronic","Pop","Rock","Hip-Hop","Electronic",
    "Pop","Rock","Hip-Hop","Electronic","Pop","Rock","Hip-Hop","Electronic",
    "Pop","Rock","Hip-Hop","Electronic"
],
"streams": [
    120,90,150,70,130,95,140,80,
    110,85,155,75,125,100,145,78,
    118,92,160,82
],
"likes": [
    30,20,50,15,35,22,48,18,
    28,19,55,16,32,25,47,17,
    29,21,60,19
],
"shares": [
    10,8,20,5,12,9,18,6,
    9,7,22,5,11,10,19,6,
    10,8,25,7
],
"year": [
    2020,2020,2020,2020,2021,2021,2021,2021,
    2022,2022,2022,2022,2023,2023,2023,2023,
    2024,2024,2024,2024
]
}

df = pd.DataFrame(data)
```

```python
genre_streams = df.groupby("genre")["streams"].sum()

plt.figure(figsize=(10, 6))
genre_streams.plot(kind="bar", color=["blue", "orange", "green", "red"])

plt.title("Total Streams by Genre")
plt.xlabel("Genre")
plt.ylabel("Total Streams")

plt.show()
```

```python
plt.figure(figsize=(10, 6))

plt.scatter(df['likes'], df['streams'], color='blue', alpha=0.6)

plt.title("Likes vs Streams")
plt.xlabel("Likes")
plt.ylabel("Streams")

plt.show()
```

```python
plt.figure(figsize=(10, 6))

plt.hist(df['streams'], bins=5, color='blue', alpha=0.7)

plt.title("Distribution of Streams")
plt.xlabel("Streams")
plt.ylabel("Frequency")

plt.show()
```
